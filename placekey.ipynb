```python
header = False
for x in range(49, (len(data_jsoned) // 1000) + 1):
    responses = pk_api.lookup_placekeys(data_jsoned[x*1000:(x+1) * 1000], verbose=True)
    blight_df_placekeys = pd.read_json(json.dumps(responses), dtype={'query_id':str})
    blight_df_placekeys.to_csv("blight_placekeys.csv", columns=None, index=False, header=header, mode="a")
    header=False
```
